### Pre Processing

```python
# The number of companies hiring in respective states

df_state = df.groupby('state').count()['company_name']
df_state
```

```python
# The number of companies who are hiring "actively"

df_active = df.groupby('hiring_status').count()['company_name']
df_active
```

```python
#
data = pd.DataFrame(df)

# Grouping by state and hiring status to get the count of companies
result = data.groupby(['state', 'hiring_status'])['company_name'].nunique().unstack(fill_value=0)

# Display the result
print(result)
```

```python
result1 = data.groupby(['hiring_status','year','month'])['company_name'].nunique().unstack(fill_value=0)
result1
```

```python
result2 = data.groupby(['hiring_status','month'])['company_name'].nunique().unstack(fill_value=0)
result2
```

```python
result3 = data.groupby(['hiring_status','seniority_level'])['company_name'].nunique().unstack(fill_value=0)
result3
```

```python
result4 = data.groupby(['state','seniority_level','year'])['company_name'].nunique().unstack(fill_value=1)
result4
```

```python
result5 = data.groupby(['hiring_status','employment_type'])['company_name'].nunique().unstack(fill_value=0)
result5
```

```python
result6 = data.groupby(['job_title','employment_type','hiring_status'])['company_name'].nunique().unstack(fill_value=0)
result6
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set Seaborn style
sns.set(style="whitegrid")
```

```python
# 1. Actively Hiring and Hiring
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='hiring_status', palette='viridis')
plt.title('Actively Hiring vs Hiring')
plt.xlabel('Hiring Status')
plt.ylabel('Count')
plt.show()
```

```python
# 2. State vs Hiring Status
plt.figure(figsize=(15, 4))
sns.countplot(data=df, x='state', hue='hiring_status', palette='viridis')
plt.title('State vs Hiring Status')
plt.xlabel('State')
plt.ylabel('Count')
plt.legend(title='Hiring Status')
plt.show()
```

```python
# 3. State vs Number of Hirings
state_hiring_count = df.groupby('state')['hiring_status'].count().reset_index()
plt.figure(figsize=(15, 4))
sns.barplot(data=state_hiring_count, x='state', y='hiring_status', palette='viridis')
plt.title('State vs Number of Hirings')
plt.xlabel('State')
plt.ylabel('Number of Hirings')
plt.show()
```

```python
# 4. Hiring Status vs Year
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='year', hue='hiring_status', palette='viridis')
plt.title('Hiring Status vs Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Hiring Status')
plt.show()
```

```python
# 5. Hiring Status vs Month
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x='month', hue='hiring_status', palette='viridis')
plt.title('Hiring Status vs Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.legend(title='Hiring Status')
plt.show()
```

```python
# 6. Hiring Status vs Seniority Level
plt.figure(figsize=(15, 4))
sns.countplot(data=df, x='seniority_level', hue='hiring_status', palette='viridis')
plt.title('Hiring Status vs Seniority Level')
plt.xlabel('Seniority Level')
plt.ylabel('Count')
plt.legend(title='Hiring Status')
plt.show()
```

```python
# 7. Employment Type vs Hiring Status
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='employment_type', hue='hiring_status', palette='viridis')
plt.title('Employment Type vs Hiring Status')
plt.xlabel('Employment Type')
plt.ylabel('Count')
plt.legend(title='Hiring Status')
plt.show()
```
